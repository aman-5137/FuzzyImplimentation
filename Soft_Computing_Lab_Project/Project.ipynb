```python
df['BP _Systolic (mmHg)'].value_counts()
```

```python
colors = ['blue', 'green', 'red', 'orange']
df['BP _Systolic (mmHg)'].value_counts().plot(kind='bar', figsize = (8, 4), color=colors)
plt.title('Different PCOS cases at different systolic sizes and Endometrium sizes')
plt.xlabel('BP _Systolic (mmHg)')
plt.ylabel('Endometrium (mm)')
plt.legend()
plt.show()
```

```python
df['Follicle No. (L)'].value_counts()
```

```python
df['Hair loss(Y/N)'].value_counts()
```

#### Estimating correlation coefficient

```python
df = df.select_dtypes(include=[int, float])
correlation = df.corr()
print(correlation)
```

#### Discover pattern and relationships

```python
# Correlation Heat Map
plt.figure(figsize=(16,12))
plt.title('Correlation of Attributes with PCOS')
a = sns.heatmap(correlation, square=True, annot=True, fmt='.1f', linecolor='white')
a.set_xticklabels(a.get_xticklabels(), rotation = 90)
a.set_yticklabels(a.get_yticklabels(), rotation = 30)
plt.show()
```

#### Pair plot

```python
num_var = ['Follicle No. (L)','Follicle No. (R)', 'Endometrium (mm)']
sns.pairplot(df[num_var], kind = 'scatter', diag_kind = 'hist')
plt.show()
```

```python
sns.lmplot(x = 'Follicle No. (R)', y = 'Follicle No. (L)', data = df)
plt.show()
```

```python
sns.lmplot(x = 'Weight (Kg)', y = 'Follicle No. (R)', data= df)
plt.title("Scatter Plot of Body Weight and Follicle")
plt.show()
```

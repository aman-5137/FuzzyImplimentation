```python
import numpy as np
import matplotlib.pyplot as plt

# Function to create triangular membership function
def triangular(x, a, b, c):
    return np.maximum(np.minimum((x - a) / (b - a), (c - x) / (c - b)), 0)

# Function to create trapezoidal membership function
def trapezoidal(x, a, b, c, d):
    return np.maximum(np.minimum((x - a) / (b - a), 1, (d - x) / (d - c)), 0)

# Range of values for age, BMI, and FSH
x_age = np.linspace(15, 50, 500)
x_bmi = np.linspace(10, 50, 500)
x_fsh = np.linspace(0, 20, 500)

# Fuzzy sets for Age
age_young = triangular(x_age, 18, 25, 30)
age_middle = triangular(x_age, 25, 35, 40)
age_old = triangular(x_age, 35, 45, 50)

# Fuzzy sets for BMI
bmi_underweight = trapezoidal(x_bmi, 0, 0, 18.5, 19)
bmi_normal = triangular(x_bmi, 18.5, 22, 24.9)
bmi_overweight = triangular(x_bmi, 25, 27.5, 29.9)
bmi_obese = trapezoidal(x_bmi, 30, 32, 40, 50)

# Fuzzy sets for FSH
fsh_low = trapezoidal(x_fsh, 0, 0, 3, 5)
fsh_normal = triangular(x_fsh, 4.5, 6.5, 8.5)
fsh_high = trapezoidal(x_fsh, 8, 10, 15, 20)

# Plotting the fuzzy sets
fig, axes = plt.subplots(3, 1, figsize=(5, 10))
axes[0].plot(x_age, age_young, label="Young")
axes[0].plot(x_age, age_middle, label="Middle-aged")
axes[0].plot(x_age, age_old, label="Old")
axes[0].set_title("Fuzzy Sets for Age")
axes[0].legend()

axes[1].plot(x_bmi, bmi_underweight, label="Underweight")
axes[1].plot(x_bmi, bmi_normal, label="Normal")
axes[1].plot(x_bmi, bmi_overweight, label="Overweight")
axes[1].plot(x_bmi, bmi_obese, label="Obese")
axes[1].set_title("Fuzzy Sets for BMI")
axes[1].legend()

axes[2].plot(x_fsh, fsh_low, label="Low")
axes[2].plot(x_fsh, fsh_normal, label="Normal")
axes[2].plot(x_fsh, fsh_high, label="High")
axes[2].set_title("Fuzzy Sets for FSH")
axes[2].legend()

plt.tight_layout()
plt.show()
```

# Implementing a fuzzy logic model

```python
import numpy as np
import pandas as pd
import skfuzzy as fuzz
from skfuzzy import control as ctrl

# Load and clean data
data = pd.read_csv(r"C:\Users\91912\OneDrive\Documents\GALGOTIAS\PCOS.csv")

# Selecting relevant columns and handling missing data
data['BMI'] = pd.to_numeric(data['BMI'], errors='coerce')
data['FSH/LH'] = pd.to_numeric(data['FSH/LH'], errors='coerce')
data['Age'] = data[' Age (yrs)']

# Drop rows with missing values for simplicity
clean_data = data[['BMI', 'FSH/LH', 'Age', 'PCOS (Y/N)']].dropna()

# Define fuzzy variables
bmi = ctrl.Antecedent(np.arange(10, 50, 1), 'BMI')
fsh_lh = ctrl.Antecedent(np.arange(0, 10, 0.1), 'FSH/LH')
age = ctrl.Antecedent(np.arange(10, 50, 1), 'Age')
pcos_risk = ctrl.Consequent(np.arange(0, 100, 1), 'PCOS Risk')

# Membership functions for BMI
bmi['low'] = fuzz.trapmf(bmi.universe, [10, 10, 18.5, 24.9])
bmi['normal'] = fuzz.trimf(bmi.universe, [18.5, 24.9, 30])
bmi['high'] = fuzz.trapmf(bmi.universe, [24.9, 30, 50, 50])

# Membership functions for FSH/LH ratio
fsh_lh['low'] = fuzz.trapmf(fsh_lh.universe, [0, 0, 1.5, 2])
fsh_lh['normal'] = fuzz.trimf(fsh_lh.universe, [1.5, 2.5, 3.5])
fsh_lh['high'] = fuzz.trapmf(fsh_lh.universe, [2.5, 3.5, 10, 10])

# Membership functions for Age
age['young'] = fuzz.trapmf(age.universe, [10, 10, 20, 30])
age['middle'] = fuzz.trimf(age.universe, [20, 30, 40])
age['old'] = fuzz.trapmf(age.universe, [30, 40, 50, 50])

# Membership functions for PCOS Risk
pcos_risk['low'] = fuzz.trapmf(pcos_risk.universe, [0, 0, 25, 50])
pcos_risk['moderate'] = fuzz.trimf(pcos_risk.universe, [25, 50, 75])
pcos_risk['high'] = fuzz.trapmf(pcos_risk.universe, [50, 75, 100, 100])

# Define rules
rule1 = ctrl.Rule(bmi['high'] & fsh_lh['high'], pcos_risk['high'])
rule2 = ctrl.Rule(bmi['normal'] & age['middle'], pcos_risk['moderate'])
rule3 = ctrl.Rule(bmi['low'] & age['young'], pcos_risk['low'])

# Create control system
pcos_ctrl = ctrl.ControlSystem([rule1, rule2, rule3])
pcos_sim = ctrl.ControlSystemSimulation(pcos_ctrl)


```
